# 3.2. Narrowband Random Signals

A stationary random process $\xi(t)$ with zero mathematical expectation is called **narrowband** if the angular bandwidth $\Delta \omega$ of the frequency region where the spectral density $S_\xi(\omega)$ is practically nonzero is small compared to some central angular frequency $\omega_0$ of this region, i.e.,

$$
S_\xi(\omega) \neq 0 \quad \text{for} \quad \omega_0 - \Delta \omega / 2 < \omega < \omega_0 + \Delta \omega / 2, \qquad \Delta \omega \ll \omega_0,
\tag{3.6}
$$

where $S_\xi(\omega)$ denotes the one-sided power spectral density (per rad/s), and $\Delta \omega$ is the effective angular bandwidth, which can be defined in various ways (e.g., 3-dB bandwidth, RMS bandwidth, or equivalent noise bandwidth).

If the one-sided spectral density $S_\xi(\omega)$ is symmetric with respect to some angular frequency $\omega_0$, then it is natural to select this frequency as the central frequency. In other cases, $\omega_0$ may be chosen more or less arbitrarily. A common choice is the spectral centroid (i.e., the first moment of the normalized spectral density):

$$
\omega_0 = \left( \int_0^{\infty} S_\xi(\omega) \, d\omega \right)^{-1}
\int_0^{\infty} \omega S_\xi(\omega) \, d\omega.
\tag{3.7}
$$

The ACF of a narrowband random process can always be represented in the form [4]:

$$
R_\xi(\tau) = \sigma_\xi^2 r_c(\tau) \cos(\omega_0 \tau) - \sigma_\xi^2 r_s(\tau) \sin(\omega_0 \tau)
= \sigma_\xi^2 r(\tau) \cos\bigl(\omega_0 \tau + \gamma(\tau)\bigr),
\tag{3.8}
$$

where

$$
r(\tau) = \sqrt{r_c^2(\tau) + r_s^2(\tau)},
\qquad
\tan \gamma(\tau) = \frac{r_s(\tau)}{r_c(\tau)},
\tag{3.9a}
$$

$$
\sigma_\xi^2 r_c(\tau) = \frac{1}{2\pi} \int_{-\omega_0}^{\infty} S_\xi(\omega_0 + \nu) \cos(\nu \tau) \, d\nu,
\tag{3.9b}
$$

$$
\sigma_\xi^2 r_s(\tau) = \frac{1}{2\pi} \int_{-\omega_0}^{\infty} S_\xi(\omega_0 + \nu) \sin(\nu \tau) \, d\nu,
\tag{3.9c}
$$

with the initial conditions

$$
r_c(0) = 1, \qquad r_s(0) = 0, \qquad \gamma(0) = 0.
\tag{3.10}
$$

Here, $\sigma_\xi^2 = R_\xi(0)$ denotes the variance of the narrowband process $\xi(t)$.

The functions $r_c(\tau)$ and $r_s(\tau)$ are respectively the in-phase and quadrature correlation components. The envelope $r(\tau)$ and phase modulation $\gamma(\tau)$ characterize the slow variations of the narrowband process relative to the carrier frequency $\omega_0$.

In the particular case where the spectral density $S_\xi(\omega)$ is symmetric with respect to the central angular frequency $\omega_0$, the ACF of the narrowband process simplifies to

$$
R_\xi(\tau) = \sigma_\xi^2 r(\tau) \cos(\omega_0 \tau)
\simeq \frac{\cos(\omega_0 \tau)}{2\pi} \int_{-\infty}^{\infty} S_\xi(\omega_0 + \nu) \cos(\nu \tau) \, d\nu,
\tag{3.11}
$$

since

$$
r_s(\tau) = 0, \qquad \gamma(\tau) = 0.
\tag{3.12}
$$

Because the right-hand sides of Equations (3.9b) and (3.9c) involve the spectral density of the process $\xi(t)$ shifted to baseband (i.e., the low-frequency spectrum), the functions $r_c(\tau)$, $r_s(\tau)$, as well as $r(\tau)$ and $\gamma(\tau)$, vary slowly compared to $\cos(\omega_0 \tau)$.

Sample functions (realizations) of a narrowband process $\xi(t)$ exhibit the structure of a modulated harmonic oscillation. Consequently, narrowband processes are commonly represented in the canonical form

$$
\xi(t) = A(t) \cos\bigl(\omega_0 t + \varphi(t)\bigr),
\qquad
A(t) \ge 0,
\qquad
-\pi < \varphi(t) \le \pi,
\tag{3.13}
$$

where $A(t)$ and $\varphi(t)$ are slowly varying functions compared to $\cos(\omega_0 t)$, termed the envelope and phase of the narrowband process $\xi(t)$, respectively.

It is also useful to introduce the concept of instantaneous angular frequency, defined as

$$
\omega(t) = \omega_0 + \dot{\varphi}(t),
\tag{3.14}
$$

where the overdot denotes differentiation with respect to time.

The representation of the narrowband process in Equation (3.13) can alternatively be expressed as

$$
\xi(t) = A_c(t) \cos(\omega_0 t) - A_s(t) \sin(\omega_0 t),
\tag{3.15}
$$

where

$$
A_c(t) = A(t) \cos \varphi(t),
\qquad
A_s(t) = A(t) \sin \varphi(t).
\tag{3.16}
$$

Consequently, it follows that

$$
A(t) = \sqrt{A_c^2(t) + A_s^2(t)},
\qquad
\varphi(t) = \arctan\!\left(\frac{A_s(t)}{A_c(t)}\right),
\tag{3.17}
$$

$$
\omega(t) = \omega_0 + \frac{\dot{A}_s(t) A_c(t) - \dot{A}_c(t) A_s(t)}{A_c^2(t) + A_s^2(t)}.
\tag{3.18}
$$

The representations in Equations (3.13) and (3.15) admit a geometric interpretation: the envelope $A(t)$ is the magnitude of a vector whose projections onto the axes of a rectangular coordinate system are $A_c(t)$ and $A_s(t)$, and the phase $\varphi(t)$ is the angle subtended between the abscissa axis and the vector direction. Both $A(t)$ and $\varphi(t)$ vary randomly in time, causing the tip of the vector to execute a random walk in the plane; the narrowband process $\xi(t)$ can thus be visualized as a rotating vector with randomly fluctuating amplitude and phase. The projections $A_c(t)$ and $A_s(t)$ are accordingly termed the quadrature components of the process $\xi(t)$.

# Task

Show that if $\xi(t)$ is ruled by a Gaussian distribution, then

- $A_c(t)$ and $A_s(t)$ are also Gaussian-distributed, each one having variance $\sigma_c^2 = \sigma_s^2 = \sigma_\xi^2$.
- $A_c(t)$ and $A_s(t)$ are zero-mean, if $\xi(t)$ is zero mean.
- $A_c(t)$ and $A_s(t)$ are zero-correlated regardless of the lag $\tau$, if $S_\xi(\omega)$ is symmetric with respect to $\omega_0$.

## Worked Solution

A standard way to obtain the quadrature components is by coherent demodulation followed by ideal low-pass filtering:

$$
A_c(t) = \operatorname{LPF}\!\left\{ 2\,\xi(t)\cos(\omega_0 t) \right\},
\qquad
A_s(t) = \operatorname{LPF}\!\left\{ -2\,\xi(t)\sin(\omega_0 t) \right\}.
$$

These operations are linear: multiplication by a deterministic sinusoid is linear, and ideal low-pass filtering is also linear. Therefore, if $\xi(t)$ is Gaussian, then $A_c(t)$ and $A_s(t)$ are Gaussian as well, because linear transformations preserve Gaussianity.

To determine their second-order properties, define

$$
R_{cc}(\tau) = \mathbb{E}\{A_c(t)A_c(t+\tau)\},
\qquad
R_{ss}(\tau) = \mathbb{E}\{A_s(t)A_s(t+\tau)\},
$$

$$
R_{cs}(\tau) = \mathbb{E}\{A_c(t)A_s(t+\tau)\},
\qquad
R_{sc}(\tau) = \mathbb{E}\{A_s(t)A_c(t+\tau)\}.
$$

Using Equation (3.15),

$$
\xi(t) = A_c(t)\cos(\omega_0 t) - A_s(t)\sin(\omega_0 t).
$$

### 1. Gaussianity and Variance of $A_c(t)$ and $A_s(t)$

Expanding the autocorrelation of $\xi(t)$ gives

$$
R_\xi(\tau) = \mathbb{E}\{\xi(t)\xi(t+\tau)\}
$$

$$
= \frac{R_{cc}(\tau) + R_{ss}(\tau)}{2}\cos(\omega_0\tau)
+ \frac{R_{cc}(\tau) - R_{ss}(\tau)}{2}\cos(2\omega_0 t + \omega_0\tau)
$$

$$
- \frac{R_{cs}(\tau) + R_{sc}(\tau)}{2}\sin(2\omega_0 t + \omega_0\tau)
- \frac{R_{cs}(\tau) - R_{sc}(\tau)}{2}\sin(\omega_0\tau).
$$

Since $\xi(t)$ is stationary, $R_\xi(\tau)$ cannot depend on the absolute time $t$. Hence the coefficients of the terms containing $2\omega_0 t$ must vanish:

$$
R_{cc}(\tau) = R_{ss}(\tau),
\qquad
R_{sc}(\tau) = -R_{cs}(\tau).
$$

Therefore,

$$
R_\xi(\tau) = R_{cc}(\tau)\cos(\omega_0\tau) - R_{cs}(\tau)\sin(\omega_0\tau).
$$

Comparing this with Equation (3.8),

$$
R_\xi(\tau) = \sigma_\xi^2 r_c(\tau)\cos(\omega_0\tau) - \sigma_\xi^2 r_s(\tau)\sin(\omega_0\tau),
$$

we identify

$$
R_{cc}(\tau) = R_{ss}(\tau) = \sigma_\xi^2 r_c(\tau),
\qquad
R_{cs}(\tau) = \sigma_\xi^2 r_s(\tau).
$$

Now evaluate at $\tau = 0$. Since Equation (3.10) gives $r_c(0)=1$,

$$
\sigma_c^2 = R_{cc}(0) = \sigma_\xi^2 r_c(0) = \sigma_\xi^2,
$$

$$
\sigma_s^2 = R_{ss}(0) = \sigma_\xi^2 r_c(0) = \sigma_\xi^2.
$$

So both quadrature components are Gaussian-distributed and each has variance equal to the variance of $\xi(t)$.

### 2. Zero-Mean Property

Because $A_c(t)$ and $A_s(t)$ are obtained from $\xi(t)$ by linear operations, their means are linear functions of the mean of $\xi(t)$. If $\xi(t)$ is zero mean, then

$$
\mathbb{E}\{A_c(t)\} = 0,
\qquad
\mathbb{E}\{A_s(t)\} = 0.
$$

The same conclusion follows directly from Equation (3.15):

$$
\mathbb{E}\{\xi(t)\} = \mathbb{E}\{A_c(t)\}\cos(\omega_0 t) - \mathbb{E}\{A_s(t)\}\sin(\omega_0 t).
$$

If $\mathbb{E}\{\xi(t)\}=0$ for every $t$, the only way this identity can hold for all $t$ is

$$
\mathbb{E}\{A_c(t)\} = 0,
\qquad
\mathbb{E}\{A_s(t)\} = 0.
$$

Hence both quadrature components are zero mean whenever the original narrowband Gaussian process is zero mean.

### 3. Zero Correlation When $S_\xi(\omega)$ Is Symmetric About $\omega_0$

From Equation (3.9c),

$$
\sigma_\xi^2 r_s(\tau) = \frac{1}{2\pi}\int_{-\omega_0}^{\infty} S_\xi(\omega_0 + \nu)\sin(\nu\tau)\,d\nu.
$$

If the spectrum is symmetric with respect to $\omega_0$, then the shifted spectrum $S_\xi(\omega_0+\nu)$ is an even function of $\nu$, while $\sin(\nu\tau)$ is an odd function of $\nu$. Their product is odd, so the integral cancels. This is exactly the special case stated in Equation (3.12):

$$
r_s(\tau) = 0.
$$

Using the identification found above,

$$
R_{cs}(\tau) = \sigma_\xi^2 r_s(\tau) = 0 \qquad \text{for all } \tau.
$$

Therefore $A_c(t)$ and $A_s(t)$ are uncorrelated for every lag whenever the narrowband spectrum is symmetric around the carrier frequency $\omega_0$.

## Final Conclusions

If $\xi(t)$ is a narrowband Gaussian random process, then:

1. $A_c(t)$ and $A_s(t)$ are Gaussian because they are obtained from $\xi(t)$ through linear operations.
2. Their variances are

$$
\sigma_c^2 = \sigma_s^2 = \sigma_\xi^2.
$$

3. If $\xi(t)$ is zero mean, then

$$
\mathbb{E}\{A_c(t)\} = \mathbb{E}\{A_s(t)\} = 0.
$$

4. If $S_\xi(\omega)$ is symmetric about $\omega_0$, then

$$
R_{cs}(\tau) = 0 \qquad \forall\tau,
$$

so the two quadrature components are zero-correlated for all lags.

## Empirical Verification

The following experiment builds a symmetric narrowband Gaussian process numerically and then recovers its quadrature components by coherent demodulation:

$$
\xi(t) = A_c(t)\cos(\omega_0 t) - A_s(t)\sin(\omega_0 t).
$$

Here, $A_c(t)$ and $A_s(t)$ are generated as independent zero-mean Gaussian low-pass processes with the same bandwidth. This construction gives a narrowband Gaussian signal with a spectrum symmetric around the carrier frequency. After that, we estimate

$$
\widehat{A}_c(t) = \operatorname{LPF}\!\left\{2\xi(t)\cos(\omega_0 t)\right\},
\qquad
\widehat{A}_s(t) = \operatorname{LPF}\!\left\{-2\xi(t)\sin(\omega_0 t)\right\},
$$

and check numerically that:

1. $\widehat{A}_c(t)$ and $\widehat{A}_s(t)$ look Gaussian.
2. Their sample means are near zero and their sample variances are close to the variance of $\xi(t)$.
3. Their empirical cross-correlation is close to zero for all small lags.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(7)


def fft_lowpass(x, fs, cutoff_hz, order=12):
    x = np.asarray(x, dtype=float)
    freqs = np.fft.rfftfreq(x.size, d=1.0 / fs)
    response = 1.0 / np.sqrt(1.0 + (freqs / cutoff_hz) ** (2 * order))
    return np.fft.irfft(np.fft.rfft(x) * response, n=x.size)


def generate_baseband_gaussian(n, fs, cutoff_hz, rng, order=12):
    white = rng.standard_normal(n)
    shaped = fft_lowpass(white, fs, cutoff_hz=cutoff_hz, order=order)
    shaped = shaped - np.mean(shaped)
    shaped = shaped / np.std(shaped, ddof=0)
    return shaped


def gaussian_pdf(x, mean, std):
    z = (x - mean) / std
    return np.exp(-0.5 * z**2) / (std * np.sqrt(2.0 * np.pi))


def standardized_moments(x):
    x = np.asarray(x, dtype=float)
    z = (x - np.mean(x)) / np.std(x, ddof=0)
    skewness = np.mean(z**3)
    excess_kurtosis = np.mean(z**4) - 3.0
    return float(skewness), float(excess_kurtosis)


def normalized_crosscorr(x, y, max_lag):
    x = np.asarray(x, dtype=float) - np.mean(x)
    y = np.asarray(y, dtype=float) - np.mean(y)
    n = x.size
    nfft = 1 << int(np.ceil(np.log2(2 * n - 1)))
    corr = np.fft.ifft(np.fft.fft(x, nfft) * np.conj(np.fft.fft(y, nfft))).real
    corr = np.concatenate((corr[-(n - 1):], corr[:n]))
    lags = np.arange(-(n - 1), n)
    keep = np.abs(lags) <= max_lag
    norm = np.sqrt(np.dot(x, x) * np.dot(y, y))
    return lags[keep], corr[keep] / norm


def one_sided_psd(x, fs):
    x = np.asarray(x, dtype=float) - np.mean(x)
    window = np.hanning(x.size)
    spectrum = np.fft.rfft(x * window)
    psd = np.abs(spectrum) ** 2 / (fs * np.sum(window**2))
    if x.size % 2 == 0:
        psd[1:-1] *= 2.0
    else:
        psd[1:] *= 2.0
    freq = np.fft.rfftfreq(x.size, d=1.0 / fs)
    return freq, psd
```

```python
fs = 4000.0
duration_s = 20.0
n = int(fs * duration_s)
t = np.arange(n) / fs

carrier_hz = 350.0
baseband_cutoff_hz = 18.0
recovery_cutoff_hz = 40.0
max_lag_samples = 250
edge_trim = int(1.0 * fs)

Ac_true = generate_baseband_gaussian(n, fs, baseband_cutoff_hz, rng)
As_true = generate_baseband_gaussian(n, fs, baseband_cutoff_hz, rng)

carrier_cos = np.cos(2.0 * np.pi * carrier_hz * t)
carrier_sin = np.sin(2.0 * np.pi * carrier_hz * t)
xi = Ac_true * carrier_cos - As_true * carrier_sin

Ac_hat = fft_lowpass(2.0 * xi * carrier_cos, fs, recovery_cutoff_hz)
As_hat = fft_lowpass(-2.0 * xi * carrier_sin, fs, recovery_cutoff_hz)

def sample_summary(x):
    skewness, excess_kurtosis = standardized_moments(x)
    return {
        "mean": float(np.mean(x)),
        "variance": float(np.var(x, ddof=0)),
        "skewness": skewness,
        "excess_kurtosis": excess_kurtosis,
    }


eval_slice = slice(edge_trim, -edge_trim)
xi_eval = xi[eval_slice]
Ac_true_eval = Ac_true[eval_slice]
As_true_eval = As_true[eval_slice]
Ac_eval = Ac_hat[eval_slice]
As_eval = As_hat[eval_slice]

summary_xi = sample_summary(xi_eval)
summary_Ac = sample_summary(Ac_eval)
summary_As = sample_summary(As_eval)

corr_Ac_As = float(np.corrcoef(Ac_eval, As_eval)[0, 1])
corr_Ac_recovery = float(np.corrcoef(Ac_true_eval, Ac_eval)[0, 1])
corr_As_recovery = float(np.corrcoef(As_true_eval, As_eval)[0, 1])

lags, r_cs = normalized_crosscorr(Ac_eval, As_eval, max_lag=max_lag_samples)
lag_axis_s = lags / fs
rms_r_cs = float(np.sqrt(np.mean(r_cs**2)))

freq_xi, psd_xi = one_sided_psd(xi, fs)
offsets = np.linspace(0.0, 60.0, 300)
psd_right = np.interp(carrier_hz + offsets, freq_xi, psd_xi)
psd_left = np.interp(np.maximum(carrier_hz - offsets, 0.0), freq_xi, psd_xi)
symmetry_nmse = float(np.mean((psd_right - psd_left) ** 2) / np.mean(psd_right**2))

print("Sample summary for xi(t):")
for key, value in summary_xi.items():
    print(f"  {key:>16s}: {value:+.6f}")

print("\nSample summary for A_c(t):")
for key, value in summary_Ac.items():
    print(f"  {key:>16s}: {value:+.6f}")

print("\nSample summary for A_s(t):")
for key, value in summary_As.items():
    print(f"  {key:>16s}: {value:+.6f}")

print("\nKey empirical checks:")
print(f"  corr(A_c, A_s) at lag 0       = {corr_Ac_As:+.6f}")
print(f"  max |r_cs(tau)| for shown lags = {np.max(np.abs(r_cs)):.6f}")
print(f"  rms r_cs(tau) for shown lags   = {rms_r_cs:.6f}")
print(f"  corr(A_c true, A_c recovered)  = {corr_Ac_recovery:+.6f}")
print(f"  corr(A_s true, A_s recovered)  = {corr_As_recovery:+.6f}")
print(f"  PSD symmetry NMSE around f0    = {symmetry_nmse:.6e}")

assert abs(summary_Ac["mean"]) < 0.05
assert abs(summary_As["mean"]) < 0.05
assert abs(summary_Ac["variance"] - summary_xi["variance"]) < 0.12
assert abs(summary_As["variance"] - summary_xi["variance"]) < 0.12
assert abs(corr_Ac_As) < 0.15
assert rms_r_cs < 0.08

excerpt = int(0.20 * fs)
hist_axis = np.linspace(-4.0, 4.0, 500)
freq_keep = (freq_xi >= carrier_hz - 80.0) & (freq_xi <= carrier_hz + 80.0)

fig, axes = plt.subplots(2, 2, figsize=(13, 8))

axes[0, 0].plot(t[:excerpt], xi[:excerpt], lw=1.0)
axes[0, 0].set_title("Narrowband Gaussian signal excerpt")
axes[0, 0].set_xlabel("time [s]")
axes[0, 0].set_ylabel(r"$\xi(t)$")

axes[0, 1].plot(freq_xi[freq_keep], psd_xi[freq_keep], lw=1.8)
axes[0, 1].axvline(carrier_hz, color="tab:red", ls="--", lw=1.2)
axes[0, 1].set_title("PSD around the carrier")
axes[0, 1].set_xlabel("frequency [Hz]")
axes[0, 1].set_ylabel(r"$S_\xi(f)$")

axes[1, 0].hist(Ac_hat, bins=80, density=True, alpha=0.65, label=r"$\hat{A}_c(t)$")
axes[1, 0].plot(
    hist_axis,
    gaussian_pdf(hist_axis, np.mean(Ac_hat), np.std(Ac_hat, ddof=0)),
    color="black",
    lw=2,
    label="Gaussian fit",
)
axes[1, 0].set_title(r"Histogram of $\hat{A}_c(t)$")
axes[1, 0].set_xlabel(r"$\hat{A}_c$")
axes[1, 0].set_ylabel("density")
axes[1, 0].legend()

axes[1, 1].plot(lag_axis_s, r_cs, lw=1.8)
axes[1, 1].axhline(0.0, color="black", lw=0.8, alpha=0.5)
axes[1, 1].set_title(r"Empirical cross-correlation of $\hat{A}_c$ and $\hat{A}_s$")
axes[1, 1].set_xlabel(r"lag $\tau$ [s]")
axes[1, 1].set_ylabel(r"$r_{cs}(\tau)$")

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].hist(As_hat, bins=80, density=True, alpha=0.65, label=r"$\hat{A}_s(t)$", color="tab:orange")
axes[0].plot(
    hist_axis,
    gaussian_pdf(hist_axis, np.mean(As_hat), np.std(As_hat, ddof=0)),
    color="black",
    lw=2,
    label="Gaussian fit",
)
axes[0].set_title(r"Histogram of $\hat{A}_s(t)$")
axes[0].set_xlabel(r"$\hat{A}_s$")
axes[0].set_ylabel("density")
axes[0].legend()

axes[1].plot(t[:excerpt], Ac_true[:excerpt], label=r"$A_c$ true", lw=1.4)
axes[1].plot(t[:excerpt], Ac_hat[:excerpt], label=r"$\hat{A}_c$ recovered", lw=1.2, ls="--")
axes[1].plot(t[:excerpt], As_true[:excerpt], label=r"$A_s$ true", lw=1.4, alpha=0.85)
axes[1].plot(t[:excerpt], As_hat[:excerpt], label=r"$\hat{A}_s$ recovered", lw=1.2, ls="--", alpha=0.85)
axes[1].set_title("True vs. recovered quadrature components")
axes[1].set_xlabel("time [s]")
axes[1].set_ylabel("amplitude")
axes[1].legend(ncol=2)

plt.tight_layout()
plt.show()
```

### Interpretation of the Numerical Results

This empirical test is expected to show the same conclusions derived analytically:

- The histograms of $\hat{A}_c(t)$ and $\hat{A}_s(t)$ should closely match Gaussian bell curves, and their skewness and excess kurtosis should stay close to zero.
- Their sample means should remain close to zero.
- Their sample variances should be close to the sample variance of $\xi(t)$.
- The zero-lag correlation and the short-lag RMS cross-correlation between $\hat{A}_c(t)$ and $\hat{A}_s(t)$ should remain small.
- The PSD of $\xi(t)$ around the carrier should look approximately symmetric, which is the spectral condition used in the third result.

Because this is a finite-length simulation with estimated filtering and estimated statistics, the values are not expected to be exactly zero or exactly equal. The relevant point is that they should be close enough to the theoretical predictions to confirm the model empirically.
